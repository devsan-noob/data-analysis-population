```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'C:/Users/sanst/Downloads/API_SP.POP.TOTL_DS2_en_csv_v2_76253.csv'
df_raw = pd.read_csv(file_path, skiprows=4)

print(df_raw.head())
year_cols = [col for col in df_raw.columns if col.strip().isdigit()]
print('Year columns:', year_cols)
selected_year = '2016' if '2016' in year_cols else year_cols[-1]
print('Selected year for analysis:', selected_year)
pop_df = df_raw[['Country Name', selected_year]].dropna()
df = pop_df.rename(columns={selected_year: 'Population'})
df['Population'] = pd.to_numeric(df['Population'], errors='coerce')
df = df.dropna(subset=['Population'])

plt.figure(figsize=(9, 6))
plt.subplots_adjust(left=0.15, right=0.85, top=0.85, bottom=0.15)
plt.hist(df['Population'], bins=20, color='#766CDB', edgecolor='#333333')
plt.title('Distribution of Total Population in ' + selected_year, fontsize=20, weight='semibold', pad=15, color='#222222')
plt.xlabel('Population', fontsize=16, weight='medium', labelpad=10, color='#333333')
plt.ylabel('Number of Countries', fontsize=16, weight='medium', labelpad=10, color='#333333')
plt.xticks(fontsize=14, color='#555555')
plt.yticks(fontsize=14, color='#555555')
plt.grid(color='#E0E0E0', linestyle='--', linewidth=0.7)
plt.gca().set_axisbelow(True)
plt.show()

df_top10 = df.sort_values(by='Population', ascending=False).head(10)

plt.figure(figsize=(9, 6))
plt.subplots_adjust(left=0.15, right=0.85, top=0.85, bottom=0.15)

bars = plt.bar(df_top10['Country Name'], df_top10['Population'], color='#DA847C', edgecolor='#333333')
plt.title('Top 10 Countries by Total Population in ' + selected_year, fontsize=20, weight='semibold', pad=15, color='#222222')
plt.xlabel('Country', fontsize=16, weight='medium', labelpad=10, color='#333333')
plt.ylabel('Population', fontsize=16, weight='medium', labelpad=10, color='#333333')
plt.xticks(rotation=45, fontsize=14, color='#555555')
plt.yticks(fontsize=14, color='#555555')
plt.grid(color='#E0E0E0', linestyle='--', linewidth=0.7)
plt.gca().set_axisbelow(True)

plt.show()

print('Histogram and bar chart for the dataset created successfully.')
```
