```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
print(df.columns)  
df['Age'] = df['Age'].fillna(df['Age'].median())
df.drop(columns=[col for col in ['Cabin', 'Ticket', 'Name'] if col in df.columns], inplace=True)
if 'Sex' in df.columns:
    df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
if 'Embarked' in df.columns:
    df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
    df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
print("Data cleaning completed successfully!")
```

```python
print(df.describe())
print(df.info())
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.countplot(x='Survived', data=df, hue=None, palette='pastel', legend=False)
plt.title('Survival Count')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.xticks([0, 1], ['Male', 'Female'])
plt.title('Survival Rate by Gender')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df, hue=None, palette='muted', legend=False)
plt.title('Survival Rate by Passenger Class')
plt.show()
```
