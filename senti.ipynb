```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
try:
    df = pd.read_csv('twitter_training.csv', encoding='utf-8', header=None)
except Exception as e:
    print('Error loading dataset:', e)
    raise e
print(df.head())
df.columns = ['id', 'brand', 'sentiment', 'tweet']
print('Unique sentiments:', df['sentiment'].unique())
sentiment_counts = df['sentiment'].value_counts()
print('Sentiment counts:')
print(sentiment_counts)
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots(figsize=(9, 6))
plt.subplots_adjust(left=0.15, right=0.85, top=0.85, bottom=0.15)

sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette=['#766CDB', '#DA847C', '#D9CC8B', '#7CD9A5', '#877877', '#52515E'], ax=ax)

ax.set_title('Sentiment Distribution', pad=15, fontsize=20, fontweight='semibold', color='#222222')
ax.set_xlabel('Sentiment', labelpad=10, fontsize=16, fontweight='medium', color='#333333')
ax.set_ylabel('Count', labelpad=10, fontsize=16, fontweight='medium', color='#333333')
ax.tick_params(axis='both', which='major', labelsize=14, labelcolor='#555555')
ax.set_axisbelow(True)

plt.show()

brand_sentiment = df.groupby(['brand', 'sentiment']).size().reset_index(name='counts')
print(brand_sentiment.head())

unique_brands = df['brand'].unique()
if len(unique_brands) > 1:
    fig, ax = plt.subplots(figsize=(9, 6))
    plt.subplots_adjust(left=0.15, right=0.85, top=0.85, bottom=0.15)
    sns.barplot(data=brand_sentiment, x='brand', y='counts', hue='sentiment', ax=ax, palette=['#766CDB', '#DA847C', '#D9CC8B', '#7CD9A5', '#877877', '#52515E'])
    ax.set_title('Sentiment Distribution by Brand', pad=15, fontsize=20, fontweight='semibold', color='#222222')
    ax.set_xlabel('Brand', labelpad=10, fontsize=16, fontweight='medium', color='#333333')
    ax.set_ylabel('Count', labelpad=10, fontsize=16, fontweight='medium', color='#333333')
    plt.xticks(rotation=45, ha='right', fontsize=10, color='#555555')
    plt.yticks(fontsize=14, color='#555555')
    ax.set_axisbelow(True)
    plt.legend(title='Sentiment', fontsize=12, title_fontsize=12, loc='upper right')
    plt.show()

print('Analysis complete.')
print('done')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('punkt')

df = pd.read_csv('twitter_training.csv', header=None, encoding='utf-8')
df.columns = ['id', 'brand', 'sentiment', 'tweet']

top_brands = df['brand'].value_counts().head(10)
print("Top 10 brands by tweet count:")
print(top_brands)

plt.figure(figsize=(9, 6))
plt.subplots_adjust(left=0.15, right=0.85, top=0.85, bottom=0.15)
ax = sns.barplot(x=top_brands.index, y=top_brands.values, palette=['#766CDB', '#DA847C', '#7CD9A5', '#D9CC8B', '#877877'])
plt.title('Top 10 Brands by Tweet Count', pad=15, fontsize=20, fontweight='semibold', color='#222222')
plt.xlabel('Brand', labelpad=10, fontsize=16, fontweight='medium', color='#333333')
plt.ylabel('Count', labelpad=10, fontsize=16, fontweight='medium', color='#333333')
plt.xticks(rotation=45, ha='right', fontsize=14, color='#555555')
plt.yticks(fontsize=14, color='#555555')
ax.set_axisbelow(True)
plt.grid(color='#E0E0E0')
plt.tight_layout()
plt.show()

top5_brands = top_brands.head(5).index.tolist()
sentiment_pct = pd.DataFrame()

for brand in top5_brands:
    brand_data = df[df['brand'] == brand]
    total = len(brand_data)
    sentiment_counts = brand_data['sentiment'].value_counts()
    for sentiment in sentiment_counts.index:
        new_row = pd.DataFrame({
            'brand': [brand],
            'sentiment': [sentiment],
            'percentage': [(sentiment_counts[sentiment] / total) * 100]
        })
        sentiment_pct = pd.concat([sentiment_pct, new_row], ignore_index=True)

print("Sentiment percentages for top 5 brands:")
print(sentiment_pct.head())
```
